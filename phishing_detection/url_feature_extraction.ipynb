```python
import pandas as pd
import ipaddress
import re
import urllib.request
from bs4 import BeautifulSoup
import socket
import requests
from googlesearch import search
import whois
from datetime import date, datetime
import time
from dateutil.parser import parse as date_parse
from urllib.parse import urlparse
from feature import FeatureExtraction

df = pd.read_csv("data/phishing_verified_online.csv") #from phish tank
df.head()
df.shape

#Collecting 5000 Phishing URLs randomly
phishingUrl = df.sample(n = 5000, random_state = 24).copy()
phishingUrl = phishingUrl.reset_index(drop=True)
phishingUrl.head()

phishingUrl.shape

#Extracting the feautres & storing them in a list
phish_features = []

#converting the list to dataframe
feature_names = ['UsingIP', 'LongURL', 'ShortURL', 'Symbol@', 'Redirecting//',
                'PrefixSuffix-', 'SubDomains', 'HTTPS', 'DomainRegLen', 'Favicon',
                'NonStdPort', 'HTTPSDomainURL', 'RequestURL', 'AnchorURL',
                'LinksInScriptTags', 'ServerFormHandler', 'InfoEmail', 'AbnormalURL',
                'WebsiteForwarding', 'StatusBarCust', 'DisableRightClick',
                'UsingPopupWindow', 'IframeRedirection', 'AgeofDomain', 'DNSRecording',
                'WebsiteTraffic', 'PageRank', 'GoogleIndex', 'LinksPointingToPage',
                'StatsReport']


for i in range(0, 50):
    url = phishingUrl['url'][i]
    print(str(i) + phishingUrl['url'][i] + "\n")
    try:        
        obj = FeatureExtraction(url)
        features = obj.getFeaturesList()
        phish_features.append(features)        
        phishing = pd.DataFrame(phish_features, columns= feature_names)
        phishing.head()  
        # Storing the extracted phishing URLs fatures to csv file
        phishing.to_csv('phishing.csv', index= False)          
    except Exception as e:
        print(f"Error extracting features for URL {url}: {e}")
        continue


ldf = pd.read_csv("data/legitimate_url.csv")
ldf.head()
ldf.shape

#Collecting 5000 Phishing URLs randomly
legUrl = ldf.sample(n = 5000, random_state = 24).copy()
legUrl = legUrl.reset_index(drop=True)
legUrl.head()

legUrl.shape

#Extracting the feautres & storing them in a list
leg_features = []

#converting the list to dataframe
feature_names = ['UsingIP', 'LongURL', 'ShortURL', 'Symbol@', 'Redirecting//',
                'PrefixSuffix-', 'SubDomains', 'HTTPS', 'DomainRegLen', 'Favicon',
                'NonStdPort', 'HTTPSDomainURL', 'RequestURL', 'AnchorURL',
                'LinksInScriptTags', 'ServerFormHandler', 'InfoEmail', 'AbnormalURL',
                'WebsiteForwarding', 'StatusBarCust', 'DisableRightClick',
                'UsingPopupWindow', 'IframeRedirection', 'AgeofDomain', 'DNSRecording',
                'WebsiteTraffic', 'PageRank', 'GoogleIndex', 'LinksPointingToPage',
                'StatsReport']

for i in range(0, 50):
    url = legUrl['url'][i]
    print(str(i) + legUrl['url'][i] + "\n")
    try:
        obj = FeatureExtraction(url)
        features = obj.getFeaturesList()
        leg_features.append(features)
        legitimate = pd.DataFrame(leg_features, columns= feature_names)
        legitimate.head()  
        # Storing the extracted legitimate URLs fatures to csv file
        legitimate.to_csv('legitimate.csv', index= False)          
    except Exception as e:
        print(f"Error extracting features for URL {url}: {e}")
        continue
```
